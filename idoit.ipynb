```python
def generate_triangular_pattern(rows):
    max_width = len(str(rows * (rows + 1) // 2))
    
    for i in range(1, rows + 1):
        num = i * (i + 1) // 2  # First number in the row
        step = i  # Increment step
        
        for j in range(i):
            print(f"{num:>{max_width}}", end=" ")
            num -= step
            step -= 1
        
        print()

# Number of rows in the triangular pattern
num_rows = 4

# Generate and display the triangular pattern
generate_triangular_pattern(num_rows)

```
